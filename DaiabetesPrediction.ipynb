```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
data={
    "Glucose":[130,85,150,90,120,68,160],
    "BMI":[33.6,26.6,35.3,22.2,30.0,25.5,45.5],
    "Age":[45,29,50,31,32,49,60],
    "Outcome":[1,0,0,1,1,0,1]
}
df=pd.DataFrame(data)
print(df.head())
```

```python
x=df[["Glucose","BMI","Age"]]
y=df["Outcome"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.02,random_state=2)
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
print(y_test.values)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
newP=[[250,45.0,39]]
pred=model.predict(newP)
if pred==1:
    print("Person has diabetes")
else:
    print("Person has no diabetes")
```
