```python
# data preprocessing we convert all words to numerical here then only machine can understand(pre processing pananum)
list1=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
```

```python
df[list1]=df[list1].replace({'yes':1,'no':0})
```

```python
df.head()
```

```python
df['furnishingstatus']=df['furnishingstatus'].replace({'furnished':0,'semi-furnished':1,'unfurnished':2})
```

```python
df.head()
```

```python
#now sum column attributes like price,area this column are large number so model cant predict more accuracy so that we want to scale (standardization)
from sklearn.preprocessing import StandardScaler
scalerVariable=StandardScaler()
list2=['price','area']
#standard this variable so that model will easily identify pattern
df[list2]=scalerVariable.fit_transform(df[list2])
```

```python
#Feature Engineering and selection
x=df.drop('price',axis=1) #axis=1 means price column drop panum
y=df['price']
```

```python
x
```

```python
y
```

```python
#split data for training and testing apothan model evolo predict panuthu
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#random state means randoma 80 per for training and 20 per for testing split panum
```

```python
#model selection and training
from sklearn.linear_model import LinearRegression #oru model import panalum oru variable la load pananum
```

```python
lr=LinearRegression() #intha varaible la namba dataset la fit pananum
```

```python
lr.fit(x_train,y_train) #x_train la input iruku and y_train la output iruku
```

```python
y_pred=lr.predict(x_test) #nama model la nama test data va predict poanrom
```

```python
#test and evaluate model using metrics
from sklearn.metrics import r2_score
```

```python
lr_accuracy=r2_score(y_test,y_pred) #y_test is correct data and y_pred is nama predict panarathu (find diff betwn actual data - predict data)
lr_accuracyPercentage=lr_accuracy*100
```
